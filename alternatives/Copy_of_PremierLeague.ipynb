```python
def calculate_team_form_before_match(df, team, match_index, location, num_matches=5):
    if location == 'Home':
        team_matches = df[(df['Home_Team'] == team) & (df.index < match_index)].tail(num_matches)
        points = team_matches['Result'].apply(lambda x: 3 if x == 'H' else (1 if x == 'D' else 0)).sum()
    elif location == 'Away':
        team_matches = df[(df['Away_Team'] == team) & (df.index < match_index)].tail(num_matches)
        points = team_matches['Result'].apply(lambda x: 3 if x == 'A' else (1 if x == 'D' else 0)).sum()
    else:
        points = 0
    return points
```

```python
columns = df_cleaned.columns.tolist()
columns.remove('Result')


new_order = columns + ['Result']

df_cleaned = df_cleaned[new_order]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Assume df_cleaned is your DataFrame
# Get the counts of each class
class_counts = df_cleaned['Result'].value_counts()
min_class_count = class_counts.min()

# Downsample the majority classes
df_downsampled = pd.DataFrame()

for label in class_counts.index:
    class_subset = df_cleaned[df_cleaned['Result'] == label]

    if len(class_subset) > min_class_count:
        # Randomly sample to downsize to min_class_count
        class_subset = class_subset.sample(n=min_class_count, random_state=42)

    df_downsampled = pd.concat([df_downsampled, class_subset])

# Shuffle the downsampled DataFrame
df_downsampled = df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)


```

```python
# Shuffle the downsampled DataFrame
df_downsampled.shape
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

numerical_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('Result')

# Separate features and target
X = df_downsampled.drop(columns=['Result'])
y = df_downsampled['Result']

# Scale only numerical features
scaler = StandardScaler()
X_scaled_numerical = scaler.fit_transform(X[numerical_cols])

# Create a DataFrame for scaled numerical features
X_scaled_df = pd.DataFrame(X_scaled_numerical, columns=numerical_cols)

# Combine scaled numerical features with non-numerical features
X_non_numerical = X.drop(columns=numerical_cols)
X_final = pd.concat([X_scaled_df, X_non_numerical.reset_index(drop=True)], axis=1)

# Split the scaled data
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB


# Initialize classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(max_iter=100),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Classifier': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
}

# Store results
results = {}

# Fit classifiers and calculate accuracy
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
    disp.plot()
    plt.title(f"Confusion Matrix for {name}")
    plt.show()

    # Save accuracy and classification report
    results[name] = {
        'Accuracy': accuracy,
        'Report': classification_report(y_test, y_pred)
    }

# Print accuracy and classification report
for name, metrics in results.items():
    print(f"{name}:\nAccuracy: {metrics['Accuracy']:.2f}")
    print(f"Classification Report:\n{metrics['Report']}")
    print("=" * 50)

# Cross-validation for each classifier
for name, clf in classifiers.items():
    scores = cross_val_score(clf, X, y, cv=5)  # 5-fold cross-validation
    print(f"{name} Cross-Validation Accuracy: {scores.mean():.2f} ± {scores.std():.2f}")

# Feature importances for models that support it
importances = {}
for name, clf in classifiers.items():
    if hasattr(clf, 'feature_importances_'):
        importances[name] = clf.feature_importances_

# Visualize feature importances
for name, importance in importances.items():
    plt.bar(range(len(importance)), importance)
    plt.title(f"Feature Importances for {name}")
    plt.xlabel("Feature Index")
    plt.ylabel("Importance")
    plt.show()
```
